Visualization

```python
plt.figure(dpi=65)
sns.countplot(data['Country'])
plt.xlabel('Merkez İller')
plt.ylabel('Sayı')
plt.title('En Yoğun Satış Merkezler')
A,B,C =data.Country.value_counts()

print('İzmir -',A)
print('Aydın -',C)
print('Manisa -',B)
plt.show()
```

```python
plt.figure(dpi=75)
sns.countplot(y ='Product line', hue = "gender", data = data) 
plt.xlabel('Count')
plt.ylabel('Product Type')
plt.show()
```

```python
plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
data['gender'].value_counts().plot.pie(autopct='%1.1f%%',startangle=60)
plt.ylabel('')
plt.title("Müşteri cinsiyeti",fontsize='15')

plt.subplot(1,2,2)
data['Customer type'].value_counts().plot.pie(autopct='%1.1f%%',startangle=60)
plt.ylabel('')
plt.title("Üyelik/ Müşteri türü",fontsize='15')
plt.tight_layout()
```

```python
plt.figure(dpi=75)
sns.countplot(y ='Product line', hue = "Country", data = data) 
plt.xlabel('Count')
plt.ylabel('Product Type')
plt.show()
```

```python
plt.figure(dpi=75)
sns.countplot(x="Payment", data =data).set_title("Ödeme Seçenekleri") 
```
